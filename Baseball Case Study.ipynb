# Machin Learning Algorithms

### Decision Tree Regressor

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error,accuracy_score
from sklearn.tree import DecisionTreeRegressor
```

```python
maxacc=0
maxrm=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=i)
    dtr=DecisionTreeRegressor()
    dtr.fit(x_train,y_train)
    y_pred=dtr.predict(x_test)
    acc= accuracy_score(y_test,y_pred)
    if acc>= maxacc:
        maxacc=acc
        maxrm=i
print("Best Accuracy is = ", maxacc, "at Random forest= ", maxrm)
            
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=7)
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)    
y_pred=dtr.predict(x_test)    
print("Mean Absolute Error for Decision Tree Regression= ",mean_absolute_error(y_test,y_pred))
```

```python
for j in range(2,10):
    score=cross_val_score(dtr,x,y,cv=j)
    sic=score.mean()
    print(" At Cv= ",j)
    print("value of Cross_Val_Function = ",score)
    print(" Mean Value of Error = ", sic)
```

### Linear Regressor

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=52)
lr= LinearRegression()
lr.fit(x_train,y_train)
y_lr=lr.predict(x_test)
```

```python
print("Mean Absolute Error for Linear Regression= ",mean_absolute_error(y_test,y_lr))
```

```python
print("Mean Squared Error for Linear Regression= ",mean_squared_error(y_test,y_lr))
```

```python
for j in range(2,10):
    score=cross_val_score(lr,x,y,cv=j)
    sic=score.mean()
    print(" At Cv= ",j)
    print("value of Cross_Val_Function = ",score)
    print(" Mean Value of Error = ", sic)
```
